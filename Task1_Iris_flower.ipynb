# Model Building

spliting the Dataset

```python
X = iris.drop(['Species'], axis=1)
y = iris['Species']
print(f'X shape: {X.shape} | y shape: {y.shape} ')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0, test_size=0.33)
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

# Prediction and Performance Metric

```python
knn_predict = knn_model.predict(X_test)
```

Fitting method of the knn object

# Predict method of the knn object

```python
print(f"""
KNN Accuracy: {accuracy_score(y_test, knn_predict)}
""")
```

```python
print(f"""
KNN Confusion Matrix: 
{confusion_matrix(y_test, knn_predict)},
""")
```

```python
print(f"""
KNN Classification Report: 
{classification_report(y_test, knn_predict)},
""")
```

# Seperate Dependent and Independent Features

```python


#For Dependent feature we use 'y' variable
y=iris['Species']

#For Indenpendent feature we use 'X' variable
x=iris.drop(['Id','Species'],axis=1)
```

```python
x.shape,y.shape
```

```python
#Spliting into training and testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test =  train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier()
```

```python
classifier.fit(x_train,y_train)
```

```python
# Feature Importance
```

```python
classifier.feature_importances_
```

```python
for feature, index in zip(x, range(4)):
    print(feature,':',classifier.feature_importances_[index])
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_actual=np.array(y_test)
```

```python
print(np.round(classifier.score(x_test,y_test)*100,2),'%')
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_actual,y_pred)
```

```python
print(classification_report(y_actual,y_pred))
```
